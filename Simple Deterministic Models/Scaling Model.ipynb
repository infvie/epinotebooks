```julia
using DifferentialEquations
```

```julia
function micro_1(du, u, parms, time)
    du[1] = parms[2]*(1-u[1]/parms[4])*u[1] - parms[1]*u[2]*u[1]
    du[2] = parms[1]*u[1]*u[2]-(parms[3]+parms[5])*u[2]
end 
```

```julia
w = 1;
m = 10;
β = 0.0247*m*w^0.44;
r = 0.6*w^-0.27;
μ = 0.4*w^-0.26;
K = 16.2*w^-0.7;
α = (m-1)*μ;
```

```julia
parms = [β,r,μ,K,α];
init = [K,1.];
tspan = (0.0,10.0);
```

```julia
sir_prob = ODEProblem(micro_1,init,tspan,parms)
```

```julia
sir_sol = solve(sir_prob);
```

```julia
using Plots
```

```julia
plot(sir_sol,xlabel="Time",ylabel="Number")
```
