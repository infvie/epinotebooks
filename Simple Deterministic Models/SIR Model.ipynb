```julia
using DifferentialEquations
```

```julia
function sir_ode(du, u, p, t)
    du[1] = -p[1] * u[1] * u[2]
    du[2] = p[1] * u[1] * u[2] - p[2] * u[2]
    du[3] = p[2] * u[2]
end
```

```julia
parms = [0.1,0.05]
init = [0.99,0.01,0.0]
tspan = (0.0,200.0)
```

```julia
sir_prob = ODEProblem(sir_ode, init, tspan, parms)
```

```julia
sir_sol = solve(sir_prob, saveat = 0.1);
```

```julia
# Visualization
using Plots
plot(sir_sol,xlabel="Time",ylabel="Number")
```

```julia
function sir_ode2(du,u,p,t)
    S,I,R = u
    b,g = p
    du[1] = -b*S*I
    du[2] = b*S*I-g*I
    du[3] = g*I
end
parms = [0.1,0.05]
init = [0.99,0.01,0.0]
tspan = (0.0,200.0)
sir_prob2 = ODEProblem(sir_ode2,init,tspan,parms)
sir_sol = solve(sir_prob2,saveat = 0.1)
```
